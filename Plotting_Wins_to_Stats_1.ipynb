```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
All_Team_Stats = pd.read_csv('team_season_stat_lines.csv')
All_Team_Stats_Copy = All_Team_Stats.copy()
All_Team_Stats_df = pd.DataFrame(All_Team_Stats_Copy)
All_Team_Stats_df
```

```python
All_Team_Stats_df['team'].replace('NOH', 'NOP', inplace=True)
All_Team_Stats_df
```

```python
All_Team_Stats_df.columns
```

```python

def plot_wins_and_stats(year, stat1, stat2):
    # Filter data for the specified year
    df = All_Team_Stats_df[All_Team_Stats_df['season'] == year].copy()
    
    # Sort the DataFrame by wins in descending order
    df = df.sort_values(by='wins', ascending=False)
    
    # Calculate average statistics for the season
    avg_stat1 = df[stat1].mean()
    avg_stat2 = df[stat2].mean()
    avg_wins = 41  # Assuming average wins to be 41
    
    # Create subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 14))
    
    # Plot Regular Season Wins as a bar chart with line to average
    for team, wins in zip(df['team'], df['wins']):
        color = 'green' if wins > avg_wins else 'red'
        ax1.bar(team, wins, color=color)
    ax1.axhline(y=avg_wins, color='red', linestyle='dotted', label='41 Wins')
    ax1.set_ylabel('Regular Season Wins', fontsize='large')
    ax1.set_title(f'Regular Season Wins for All Teams in {year}')
    ax1.legend(fontsize='large')
    
    # Plot Stat 1
    for team, value in zip(df['team'], df[stat1]):
        color = 'green' if value > avg_stat1 else 'red'
        ax2.plot(team, value, color=color, marker='d', markersize=8)
        ax2.plot([team, team], [value, avg_stat1], color='black', linestyle='-', linewidth=0.75)
    ax2.axhline(y=avg_stat1, color='red', linestyle='dotted',
                label=f'Average {stat1} ({avg_stat1:.2f})')
    ax2.set_ylabel(f'{stat1}', fontsize='large')
    ax2.set_title(f'{stat1} for All Teams in {year}')
    ax2.legend(fontsize='large')
    
    # Plot Stat 2
    for team, value in zip(df['team'], df[stat2]):
        color = 'green' if value > avg_stat2 else 'red'
        ax3.plot(team, value, color=color, marker='d', markersize=8)
        ax3.plot([team, team], [value, avg_stat2], color='black', linestyle='-', linewidth=0.75)
    ax3.axhline(y=avg_stat2, color='red', linestyle='dotted',
                label=f'Average {stat2} ({avg_stat2:.2f})')
    ax3.set_ylabel(f'{stat2}', fontsize='large')
    ax3.set_title(f'{stat2} for All Teams in {year}')
    ax3.legend(fontsize='large')
    
    # Rotate x-axis labels for better readability
    for ax in (ax1, ax2, ax3):
        ax.tick_params(axis='x', rotation=30, labelsize=15)
        ax.tick_params(axis='y', labelsize=15)
    
    plt.tight_layout()
    plt.show()
```

```python
plot_wins_and_stats(2016, 'pts', 'three_point')
```

```python
plot_wins_and_stats(2016, 'blk', 'stl')
```
