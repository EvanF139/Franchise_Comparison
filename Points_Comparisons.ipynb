```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
All_Team_Stats = pd.read_csv('team_season_stat_lines.csv')
All_Team_Stats_df = pd.DataFrame(All_Team_Stats)
All_Team_Stats_df
```

```python
print(All_Team_Stats_df.dtypes)
```

```python
df = All_Team_Stats_df.loc[All_Team_Stats_df['team'] == 'NYK']
fig, ax1 = plt.subplots(figsize=(12, 5))
# ax1.plot(df['season'], df['salary']/1000000, color='green', linestyle='dashed', marker='s', label= 'Payroll in $Millions')
ax1.plot(df['season'], df['wins'], color='blue', linestyle='-', marker='o', label= 'Wins')
ax1.plot(df['season'], df['pts'], color='orange', linestyle='dotted', marker='^', label= 'Points per Game')
ax1.set_xlabel('NBA Season')
# ax1.set_ylabel('Value', labelpad=20, fontsize='large')
ax1.set_title('Points to Wins Comparison', fontsize='x-large')
# ax1.plot(df['season'], [41] * len(df), linestyle='dotted', color='orange', marker= ',', label='41 Wins (.500 Record)')
ax1.legend(fontsize='x-small')
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.grid(axis='y')
plt.xticks(df['season'], rotation=45)
plt.tight_layout()
plt.show()
```

```python
def plot_team_stats(team_abbr, All_Team_Stats_df):
    df1 = All_Team_Stats_df.loc[All_Team_Stats_df['team'] == team_abbr]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

    ax1.plot(df1['season'], df1['wins'], color='black', linestyle='-', marker='o', label= 'Wins')
    ax2.plot(df1['season'], df1['pts'], color='green', linestyle='-', marker='o', label= 'Points per Game')

    ax1.set_xlabel('NBA Season')
    ax1.set_title('Regular Season Wins', fontsize='x-large')
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    ax1.spines['left'].set_visible(False)
    ax1.set_xticks(np.arange(2000, 2025, 5))
    ax1.tick_params(rotation=45)
    ax1.set_yticks(np.arange(0, 82, 10))
    ax1.grid()

    ax2.set_xlabel('NBA Season')
    ax2.set_title('Points Per Game', fontsize='x-large')
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.spines['left'].set_visible(False)
    ax2.set_xticks(np.arange(2000, 2025, 5))
    ax2.tick_params(rotation=45)
    ax2.set_yticks(np.arange(80, 130, 5))
    ax2.grid()

    plt.tight_layout()
    plt.show()
```

```python
plot_team_stats('MIL', All_Team_Stats_df)
```

```python
plot_team_stats('CHA', All_Team_Stats_df)
```

```python
plot_team_stats('GSW', All_Team_Stats_df)
```

```python
plot_team_stats('ATL', All_Team_Stats_df)
```
